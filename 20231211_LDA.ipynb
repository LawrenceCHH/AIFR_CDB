# LDA

```python
import jieba
import pandas as pd

def get_stopwords(stopword_txt_path='/workspace/stopwords_added.txt'):
    stopwords = [line.strip() for line in open(stopword_txt_path, 'r', encoding='utf-8-sig').readlines()]
    return stopwords

def seg_jud_text(jud_text):
    jud_depart = jieba.cut(jud_text.strip())
    stopwords = get_stopwords()
    segmented = ''

    for word in jud_depart:
        if word not in stopwords:
            if word != '\t':
                segmented += word
                segmented += " "
    return segmented

target_df_path = "/workspace/20231213_fee_不復記憶&&衡.csv"
target_df_column = "fee"
target_texts = pd.read_csv(target_df_path)[target_df_column].tolist()

# -------------------------------Word segmentation-------------------------------

segmented_texts = [seg_jud_text(text).split(" ")[:-1] for text in target_texts]
print(len(segmented_texts))
print(segmented_texts[0])
```

```python
import gensim
import gensim.corpora as corpora
```

```python
# -------------------------------Bag of words-------------------------------
# 將字典轉換為詞袋,為文件中的每一個單字建立唯一的ID
id2word = corpora.Dictionary(segmented_texts)
id2word.filter_extremes(no_below=10, no_above=0.5)
corpus = [id2word.doc2bow(sentence) for sentence in segmented_texts]
# 回傳結果[[(1,3),(2,1)],[]] ，如（1,3）代表1在第一個文檔中出現3次。
print(corpus[:1])
# 查詢id對應的詞
print([[(id2word[id], freq) for id, freq in cp] for cp in corpus[:1]])
```

```python
from pprint import pprint
```

```python
# -------------------------------建立主題模型-------------------------------
# 在建立主題模型時，除了字典和語料庫之外，我們還需要調整模型的參數，如主題資料量等
# num_topics代表產生的主題數
# chunksize是每個訓練區塊中使用的文件數;
# update_every決定應更新模型參數的頻率;
# passes訓練通過的總數;

# Set training parameters.
num_topics = 10
chunksize = 2000
passes = 20
iterations = 400
# Don't evaluate model perplexity, takes too much time.
eval_every = None  

lda_model = gensim.models.ldamodel.LdaModel(
    corpus=corpus,
    id2word=id2word,
    chunksize=chunksize,
    alpha='auto',
    eta='auto',
    iterations=iterations,
    num_topics=num_topics,
    passes=passes,
    eval_every=eval_every,
    random_state=100,
)
# 上述LDA模型由10個不同的主題建立，其中每個主題是關鍵字的組合，並且每個關鍵字對主題貢獻一定的權重，權重反應了關鍵字對主體的貢獻程度。
# num_word代表每個主題的關鍵字數
pprint(lda_model.print_topics(num_words=20))
```

```python
from gensim import  models
# Smaller score has silmilar meaning
from gensim.matutils import hellinger
from sklearn.cluster import KMeans
import numpy as np
lda = models.ldamodel.LdaModel.load("/workspace/20231214_LDA_fee")

target_df_path2 = "/workspace/20231213_opinion_強暴&&難認.csv"
target_df_column2 = "opinion"
target_texts2 = pd.read_csv(target_df_path2)[target_df_column2].tolist()
segmented_texts2 = [seg_jud_text(text).split(" ")[:-1] for text in target_texts2]

# # 如果你想獲得每個主題的關鍵詞，可以使用以下方法
# topic_keywords = lda.print_topics(num_words=10)  # 設定num_words為你想要的每個主題的關鍵詞數量
# for topic in topic_keywords:
#     print(topic)

# # 比較相似度
# new_doc1 = id2word.doc2bow(segmented_texts2[0])
# topics1 = lda.get_document_topics(new_doc1)
# print(segmented_texts2[0])
# print(0, topics1, '\n')
# # topics1 = lda[new_doc1]
# scores = [100]
# for i in range(1,60):
#     # print(segmented_texts2[i])
#     new_doc2 = id2word.doc2bow(segmented_texts2[i])
#     topics2 = lda.get_document_topics(new_doc2)
#     similarity = hellinger(topics1, topics2)
#     print(i, topics2, similarity)
#     scores.append(similarity)
#     # print(similarity)
# print(min(scores))

# 製作topic vector
max_dimensions = 10  # 指定最大維度
features = []
for i in range(0,60):
    new_doc = id2word.doc2bow(segmented_texts2[i])
    topics = lda[new_doc]
    feature_vector = [0.0] * max_dimensions  # 初始化特徵向量為全零
    for topic in topics:
        topic_index = topic[0]
        topic_weight = topic[1]
        feature_vector[topic_index] = topic_weight  # 使用主題權重填充特徵向量的相應位置
    features.append(feature_vector)

# 將特徵向量轉換為NumPy數組
X = np.array(features)

# 使用K均值聚類進行分群（假設要分成3個群）
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(X)

# 獲取每個文本的所屬群
cluster_labels = kmeans.labels_

# 打印每個文本的群分配
for i, label in enumerate(cluster_labels):
    print(f"文本{i} 屬於群 {label} {features[i]}")

```

## MeanShift

```python
from sklearn.cluster import MeanShift
import numpy as np

# Create a list to store your feature vectors
max_dimensions = 10  # Specify the maximum dimensions
features = []

# Iterate through your data and create feature vectors
for i in range(0, 60):
    new_doc = id2word.doc2bow(segmented_texts2[i])
    topics = lda[new_doc]
    feature_vector = [0.0] * max_dimensions  # Initialize the feature vector with zeros
    for topic in topics:
        topic_index = topic[0]
        topic_weight = topic[1]
        feature_vector[topic_index] = topic_weight  # Fill the feature vector with topic weights
    features.append(feature_vector)

# Convert feature vectors to a NumPy array
X = np.array(features)

# Use MeanShift clustering
meanshift = MeanShift()
meanshift.fit(X)

# Get cluster labels and cluster centers
cluster_labels = meanshift.labels_
cluster_centers = meanshift.cluster_centers_

# Print the cluster labels for each data point and their corresponding features
for i, label in enumerate(cluster_labels):
    print(f"文本 {i} 屬於群 {label} {features[i]}")

# Optionally, you can also print the cluster centers
print("Cluster Centers:")
print(cluster_centers)
```

```python
from sklearn.metrics.pairwise import euclidean_distances
# Calculate distances using euclidean_distances
distances = euclidean_distances(X, cluster_centers)

# Find the index of the nearest data point for each cluster
nearest_points = np.argmin(distances, axis=0)

# Print the index of the nearest point for each cluster
for cluster_label, nearest_point_index in enumerate(nearest_points):
    print(f"Cluster {cluster_label}: Nearest data point at index {nearest_point_index}")
    print(target_texts2[nearest_point_index])
```
